```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Social_Net_class.csv")
c0=df[df["Purchased"]==0]
c1=df[df["Purchased"]==1]
```

```python
plt.scatter(c0["Age"],c0["EstimatedSalary"],c="red")
plt.scatter(c1["Age"],c1["EstimatedSalary"],c="blue")
```

```python
x=df[["Age","EstimatedSalary"]]
y=df["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(x_train)
x_tr_sc=scaler.transform(x_train)
x_ts_sc=scaler.transform(x_test)
```

```python
model.fit(x_tr_sc,y_train)
model.score(x_ts_sc,y_test)
```

```python
plt.scatter(x_test["Age"][y==0],x_test["EstimatedSalary"][y==0],c="r")
plt.scatter(x_test["Age"][y==1],x_test["EstimatedSalary"][y==1],c="b")
```

```python
ypre=model.predict(x_ts_sc)
```

```python
plt.scatter(x_test["Age"][ypre==0],x_test["EstimatedSalary"][ypre==0],c="r")
plt.scatter(x_test["Age"][ypre==1],x_test["EstimatedSalary"][ypre==1],c="b")
```
